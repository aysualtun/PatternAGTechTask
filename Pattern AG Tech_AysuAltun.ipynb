```python
df_soil
```

```python
df_soil["hzlweight"]= abs(df_soil["hzdept"]-df_soil["hzdepb"])/(df_soil["hzdepb"])
```

```python
df_soil
```

```python
df_soil.mukey.unique()
```

```python
def weighted_average(dataframe, value, weight):
    val = dataframe[value]
    wt = dataframe[weight]
    return (val * wt).sum() / wt.sum()
 
```

```python
df_soil.groupby('mukey').apply(weighted_average,
                                     'hzlweight', 'comppct')
```
